```python
y = main_file['total_claim_amount']
X = main_file.drop(['total_claim_amount'], axis=1)
X_num = X.select_dtypes(include = np.number)
X_cat1 = main_file.select_dtypes(include = np.object)
X_cat = X_cat1.drop(['customer'], axis=1)
pd.get_dummies(X_cat)
```

```python
encoder = OneHotEncoder(handle_unknown='error', drop='first')
encoder.fit(X_cat)
```

```python
encoded = encoder.transform(X_cat).toarray()
encoded
```

```python
encoded.shape
```

```python
X = np.concatenate([X_num, encoded], axis=1)

X
```

```python
#train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)

predictions  = model.predict(X_test)
```

```python
train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
r2_score(y_test, predictions), mean_squared_error(y_test, predictions, squared=False), mean_squared_error(y_test, predictions)
```
